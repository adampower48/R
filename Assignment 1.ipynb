```r
library(datasets)
```

```r
ChickWeight
```

```r
plot(ChickWeight[c("Time", "weight")], col=ChickWeight$Diet, main="Weights over Time")
legend(0, max(ChickWeight$weight),unique(ChickWeight$Diet),col=1:length(ChickWeight$Diet),pch=1, title="Diet")
```

```r
boxplot(weight~Time, data=ChickWeight[ChickWeight[,"Diet"]==1,], 
        xlab="Time", ylab="Weight", main="Diet 1")
```

```r
mean_DT = aggregate(list(Weight = ChickWeight$weight), by=list(Diet = ChickWeight$Diet, Time = ChickWeight$Time), FUN=mean)
mean_DT
```

```r
plot(Weight~Time, data=mean_DT, col=mean_DT$Diet, main="Mean Weights by Diet")
legend(0, max(mean_DT$Weight),unique(mean_DT$Diet),col=1:length(mean_DT$Diet),pch=1, title="Diet")
```

```r
boxplot(weight~Diet, data=ChickWeight[ChickWeight$Time == 20,], 
        xlab="Diet", ylab="Weight", main="Weights at Time = 20")
```
